# Informações de Países

Nesse experimento vamos desenvolver uma aplicação que vai nos retornar informações sobre países.

Será utilizado a biblioteca **Country Info** para retornar dados de países,informações ISO e estados/províncias dentro deles.

Sobre a Biblioteca a mesma foi instalada via **Anaconda Promp**.

```python
from countryinfo import CountryInfo
```

```python
br = CountryInfo('BR')
```

```python
br.capital() #Capital
```

```python
br.subregion() #Continente
```

```python
br.languages() #Língua 
```

```python
br.population() #População
```

```python
br.provinces() #Estados
```

```python
br.borders() #Fronteiras
```

```python
br.translations() #Traduções
```

Acima tivemos retorno de cada informação de forma separada, abaixo vou fazer a busca pra fins de que apresente todas as informações sobre o país.

```python
br.info() #todas as informações sobre o Brasil
```

```python
br.all() #informações sobre todos os países
```

```python
[(i.get('callingCodes'),i['name']) for i in br.all().values()] #Códigos de aréa de cada país
#cont..Da pra detectar um país com base no número de telefone.
```

Ou seja, um jeito simples de acessar dados dos países do mundo todo, como população, capital, estados, coordenadas geográficas e etc.
